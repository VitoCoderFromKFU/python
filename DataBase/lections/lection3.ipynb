**Ограничения(constraint):**

**1.на уровне записи(внутритабличные)**

sql
NOT NULL - поле обязательно не пустое

*CHECK(условие) - не проверяет NULL. Напр. CHECK(price>0), если price=null, то check=true

*DEFAULT (значение по умолчанию)

**2.внутритабличные(на уровне таблицы, колонки)**

**Ключ таблицы** - список колонок(столюцов). **Составной ключ** - ключ из несольких колонок(столбцов)

ключи:

*Уникальный ключ(UNIQUE) - значение ключа однозначно идентифицирует запись таблицы, т.е. нет двух строк с одинаковым значением уникального ключа

*первичный ключ (PRIMARY KEY) - выделенный уникальный ключ таблицы 

UNIQUE(p1) PRIMARY KEY(p1,p2)

**3.межтабличные ограничения и связи**

Родительствие таблицы <--> дочерние таблицы.		1<-->N, где N=(0,...k)

Межтабличная связь определяется при помощи пары ключей:

***Родительский ключ** (REFERENCES) - один из униклаьных ключей родительской таблицы(чаще всего первичный ключ)

***Внешний ключ** (FOREIGN KEY) - ключ дочерней таблицы

**Связь** - строка родительской таблицы связана со всеми строками дочерней таблицы, у которых значения внешнего ключа совпадают со значением родительского ключа этой строки

**Ограничение ссылочной целостности** - для каждой строки дочерней таблицы значение внешнего ключа, если оно не NULL, должна существовать строка в родительской таблице с тем же знаечнием родительского ключа

**Стратегии поддержания ссылочной целостности**

Родительская таблица								дочерняя таблица
