# язык зранимых процедур и триггеров

Хранимые процедуры - процедуры, которые хранятся как часть метаданных и для выполнения явно вызываются (` ` ` sql execute ` ` `)

Триггеры - специализированные версии хранимых процедур

    Связаны с таблицей и каким-либо событием на этой таблице(inserts, uodate , delete) и вызываются автоматически при поступлении этого события

Преимущества:

    *Модульное программирование, проектирование
    *Упрощение сопровождения(BL по СУБД)
    *уменьшение нагрузки на каналы связи

```SQL
CREATE FUNCTION Get_prod_counts (@prod varchar(20),
    @date1 date,
    @date2 date)
RETURNS decimal(15,2)
AS
    RETURN(SELECT SUM(count)
    FROM Products p, Orders o
    WHERE p.id=o.id_prod AND
          p.name=@prod AND
          o.odate BETWEEN @date1 AND @date2)

```

```SQL
CREATE FUNCTION Get_prod_orders 
    (@prod varchar(20),
    @date1 date,
    @date2 date)
RETURNS TABLE("Необязательный список полей")
AS
    RETURN(SELECT name,odate,o.count
    FROM Products p, Orders o
    WHERE p.id=o.id_prod AND
          p.name=@prod AND
          o.odate BETWEEN @date1 AND @date2)

```

```SQL
CREATE PROCEDURE Get_prod_count
    (   @prod varchar(20),
        @date1 date,
        @date2 date,
        @sum decimal(13,2) OUTPUT)
AS BEGIN 
    SELECT (@sum=sum(count) 
    FROM Products p, Orders o
    WHERE p.id=o.id_prod AND
          p.name=@prod AND
          o.odate BETWEEN @date1 AND @date2)

```

```SQL
DECLARE @sum decimal(15,2)
EXECUTE Get_prod_count('Apples', 01.01.2022, 31.12.2022,@sum OUTPUT)

```
