
```SQL
SELECT[DISTINCT | ALL] СПИСОК.КОЛОНОК
FROM ИМЯ ТАБЛ[ПСЕВДОНИМ],...
WHERE УСЛ.ОТБОРА
GROUP BY КЛЮЧГРУПП
HAVING/// УСЛ.ОТБОРА ПОЛЕ[]
ORDER BY КЛЮЧ.УПОР.
UNION [ALL] SELECT-ОПЕРАТОР
```

СЕМАНТИКА 
1)СТРОИТСЯ W1=ДЕКАРТОВО ПРОИЗВЕДЕНИЕ FROM ТАБЛ 
2) СТРОИТСЯ W2 ВЫБОРКА ИЗ W1 ПО УСЛ.ОТБОРА
3) СТРОКИ W2 СОБИРАЮТСЯ В ГРУППЫ
В ОДНУ ГРУППУ ВХОДЯТ ВСЕ СТРОКИ С ОДИНАКОВЫМ ЗНАЧЕНИЕМ КЛЮЧА ГРУППИРОВКИ И ФОРМИРУЕТСЯ ТАБЛ W3, ГДЕ ОДНА СТРОКА ДЛЯ КАЖДОЙ ГРУППЫ
4) СТРОИТСЯ W4 - ВЫБОРКА ИЗ W3 ПО УСЛ.ОТБОРА ПОСЛЕГРУПП
5) ЕСЛИ ИМЕЕТСЯ UNION, ТО СТРОИТСЯ РЕЗУЛЬТАТ ЗАПРОСА ДЛЯ SELECT-ОПЕРАТОР
6) ПОСЛЕ UNION ВСЕ СТРОКИ  ТАБЛИЦЫ УПОРЯДОЧИВАЮТСЯ ПО КЛЮЧ.УПОР
7) ОКОНЧАТЕЛЬНЫЙ РЕЗУЛЬТАТ СТРОИТЯ ПО СПЕЦИФИКАЦИИ SELECT(СПИСОК.КОЛОНОК)

```SQL
SELECT PRJNAME,
    CASE
        WHEN BUDGET < 10000 THEN "МАЛЫЙ"
        WHEN BUDGET >= 10000 AND BUDGET<20000 THEN "СРЕДНИЙ"
        WHEN BUDGET >=20000 THEN "КРУПНЫЙ"
        ELSE "ОТРИЦАТЕЛЬНО"
    END  AS  "CATEGORY",....
    FROM PROJECTS

```

```SQL
SELECT ID,ODATE,AMT,
    SUM(AMT) OVER () AS TOTALSUM,
    SUM(AMT) OVER (PARTITION BY ODATE) AS DAYSUM,
    SUM(AMT) OVER(PARTITION BY ID_C) AS CUSTSUM,
    RANK() OVER(ORDER BY AMT[,DESC]) AS RANK_BY_AMT,
    ROW.NUMBER(),
    RANK.DES() (PARTITION C_ID ORDER BY AMT);




```
