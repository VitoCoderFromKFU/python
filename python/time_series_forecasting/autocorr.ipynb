```python
import statistics as st
import matplotlib.pyplot as plt


def autocorr(x: list) -> list:
    res = []
    avg = st.mean(x)
    znam = 0
    for i in range(len(x)):
        znam += (x[i]-avg)**2
    for i in range(1, len(x)//4+1):
        temp = 0
        for j in range(0, len(x)-i):
            temp += (x[j]-avg)*(x[i+j]-avg)
        res.append(temp/znam)
    return res


#x = [225, 235, 325, 421, 448, 425, 469, 549, 565, 513, 555, 634, 641, 551, 602, 676,
     #801, 901, 1102, 1373, 1447, 1527, 1697, 2038, 2044, 1922, 2120, 2536, 2461, 2268,
     #2523, 3074, 2998, 2893.1]
x = [120.6, 113.2,  120.6, 120.7, 115.2, 115.3, 115.5,  118.5, 118.5,
     118.5, 120, 120.5, 115.9, 116.2, 117, 117.2, 117.8, 118.5, 120.5, 120.5]
res = autocorr(x)
print(res)
#res = [0.9141693984637547, 0.8113371160132821, 0.7171468252014401, 0.6510293407979221, 0.5756217649651582, 0.4798743412631363, 0.3866960403295124, 0.3147468077770703]
plt.bar([i+1 for i in range(len(res))], res)
for i in range(len(res)):
    plt.text(i+0.7,res[i],str(round(res[i],3)))
plt.show

```
