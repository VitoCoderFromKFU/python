# Выявление аномальных наблюдений по методу Ирвина

$$
\lambda_t = \frac{|y_t-y_{t-1}|}{S_y}
$$
Где
$$
S_y=\sqrt{\frac{1}{(n-1)}\sum_{t=1}^{n}(y_t-z)^2}
$$
И z
$$z=\frac{1}{n}\sum_{t=1}^{n}y_t
$$

```python
#посчитаем lambda
import math


def sr(y):
    sum = 0
    for i in range(len(y)):
        sum += y[i]
    return sum/len(y)
def sumkv(y):
    sumkv = 0
    for i in range(len(y)):
        sumkv += (y[i]-sr(y))**2
    return sumkv

def sY(y,z):
    sumk = 0
    for i in range(len(y)):
        sumk += (y[i]-sr(y))**2
    return math.sqrt(sumk/(len(y)-1))
y = [299,297,295,310,295,297,294,300,302]
print ("Среднее арифметическое =" , sr(y))
lam = []
z = sr(y)
s = sY(y,z)
for i in range(len(y)):
    lam.append((abs(y[i]-y[i-1])/s))
    print(lam[i], i)
    anom = []
for i in range(len(lam)):
    if(lam[i]>2.9):
        anom.append(i)
        print("Аномальное значение лямбда ", lam[i],"номер ",i, y[i])


disp = (sumkv(y)/len(y))
print("Дисперсия = ",disp)

#уберем аномальные значение
def changeAnom(y,anom):
    y[anom[0]] = (y[anom[0]-1] + y[anom[0]+1])/2
changeAnom(y,anom)
print("Измененный список : ",y)



lam1 = []
z1 = sr(y)
s1 = sY(y,z1)
for i in range(len(y)):
    lam1.append((abs(y[i]-y[i-1])/s1))
    print(lam1[i], i)
anom1 = []
for i in range(len(lam1)):
    if(lam1[i]>2.9):
        anom1.append(i)
        print("Аномальное значение лямбда ", lam1[i],"номер ",i, y[i])
        

disp=(sumkv(y)/len(y))
print("Дисперсия = ",disp)
```
